## Import Relevant Libaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import random

from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score

# Set matplotlib inline
%matplotlib inline
```

### Load file

```python
# import file
df = pd.read_csv('linkedin_processed.csv')
# df.head()
df.fillna(0)
```

```python
y = df['class']
X = df.drop(columns=['id','Full name','Profile url','Location','class','Industry'])

#spliting train_test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.34)

from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import f1_score,precision_score,recall_score,accuracy_score
from sklearn.svm import LinearSVC,SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier


classifiers = [
    MultinomialNB(),
    LinearSVC(),
    LogisticRegression(solver = 'lbfgs'),
    DecisionTreeClassifier(criterion="entropy")
]

for clf in classifiers:
    clf.fit(X_train, y_train)
    Y_test_hat = clf.predict(X_test)
    print("Scores for:",clf)
    print("accuracy:",accuracy_score(Y_test_hat,y_test))
    print("f1:",f1_score(Y_test_hat,y_test,average='weighted'))
    print("precision:",precision_score(Y_test_hat,y_test,average='weighted'))
    print("recall:",recall_score(Y_test_hat,y_test,average='weighted'),"\n")
```
